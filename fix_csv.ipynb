```python
import pandas as pd
```

```python
df = pd.read_csv("AAT_Veloce_Nightslogs_24A_24B_25A_25B_01 - StellarTwins_MH_-0.8.csv")
```

```python
# Combine RA columns into single string with ":" separator
df['RA'] = df['RA'].astype(str) + ':' + df['Unnamed: 2'].astype(str) + ':' + df['Unnamed: 3'].astype(str)

# Combine Dec columns into single string with ":" separator
df['Dec'] = df['Dec'].astype(str) + ':' + df['Unnamed: 5'].astype(str) + ':' + df['Unnamed: 6'].astype(str)

df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 5', 'Unnamed: 6'], inplace=True)
df.to_csv("AAT_Veloce_Nightslogs_24A_24B_25A_25B_01 - StellarTwins_MH_-0.8_fixed.csv", index=False)
```
